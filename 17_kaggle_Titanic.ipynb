```python
from IPython.display import display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 음수표현 라이브러리
plt.rcParams['axes.unicode_minus'] = False

# 경고무시
import warnings
warnings.filterwarnings("ignore")

# 매직명령어 : 시각화 결과가 노트북에 포함되도록
%matplotlib inline
```

```python
df = pd.read_csv('train.csv', index_col= 'PassengerId')
df.head(10)
```

```python
df.columns
```

```python
df.info()
```

```python
df.Survived.value_counts()
```

- Cabin은 결측값이 너무 많으니 쓰지 않는다.
- Embarked는 결측값이 2개 뿐이니 그 2개의 행을 제거한다.
- Sex와 Embarked는 Category로 변환해준다.
- Survived의 0과 1이 불균형하니 stratify

```python
df = df.dropna(subset=['Embarked'])
```

```python
df['Name'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df['Name'] = df['Name'].apply(lambda x: 'other' if x not in ["Mr", "Miss", "Mrs", "Master"] else x)
```

```python
df[['Pclass','Sex', 'Embarked', 'Name']] = df[['Pclass','Sex', 'Embarked', 'Name']].astype('category')
```

```python
data = df[['Pclass','Sex', 'Name' ,'Age', 'SibSp', 'Parch','Fare','Embarked']]
target = df[['Survived']]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target, = train_test_split(
    data, target, test_size= 0.2 , stratify= target, random_state=42)
```

```python
!pip install xgboost

import xgboost
```

```python
from sklearn.model_selection import cross_validate
from xgboost import XGBClassifier



xgb =XGBClassifier(tree_method= 'auto',
                   enable_categorical= True,
                   learning_rate= 0.1,
                   n_estimators= 300,
                   random_state= 96)

scores = cross_validate(xgb, train_input, train_target,
                        return_train_score= True, n_jobs= -1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
xgb.fit(data, target)
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test = test.dropna(subset=['Embarked'])
test['Name'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test['Name'] = test['Name'].apply(lambda x: 'other' if x not in ["Mr", "Miss", "Mrs", "Master"] else x)
test[['Pclass','Sex', 'Embarked', 'Name']] = test[['Pclass','Sex', 'Embarked', 'Name']].astype('category')
testdata = test[['Pclass','Sex','Name', 'Age', 'SibSp', 'Parch','Fare','Embarked']]
```

```python
Survived = xgb.predict(testdata)
Survived
```

```python
result_df = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': Survived})
result_df
```

```python
result_df.to_csv('predicted_results.csv', index=False)
```
